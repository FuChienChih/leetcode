### 利用『set』的特性找出所有元素。

### 利用『count』 來確認各元素出現的頻率，若只出現一次 --> ans.append()

```python
class Solution:
    def singleNumber(self, nums) :
        unique = list(set(nums)) # 讓迴圈次數減至最少
        ans = []
        for i in unique:
            if nums.count(i) == 1:
                ans.append(i)
            else :
                continue
        return ans
```
