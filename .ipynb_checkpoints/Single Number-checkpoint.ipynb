### 想法很簡單，若count == 1 ，return

```python
class Solution:
    def singleNumber(self, nums) :
        for i in nums:
            if nums.count(i) == 1:
                return i
```

### 基於上面的方法，利用『set』的特性，想辦法讓程式執行時間減少

```python
class Solution:
    def singleNumber(self, nums) :
        unique = list(set(nums)) # 讓迴圈次數減至最少
        for i in unique:
            if nums.count(i) == 1:
                return i
```
