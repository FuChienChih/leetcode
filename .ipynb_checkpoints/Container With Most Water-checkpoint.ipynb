開始的想法：p.s.這裡將裝水的兩個隔板稱為『木板』

從 最高的木板『(max(height)』開始，找出相離最遠的兩個木板(max(weight) - min(weight)) ， maximum ＝ max(距離*高 ,maximum)。

weight = 最高木板在list中的index

然後將最高的所有木板削短成跟第二高的木板一樣高。這樣第二高的木板就變成最高的木板了～

重複此步驟，直到最高木板的高度 = min(height)

### 一開始自己的答案 (Runtime Error)

```python
import numpy as np
class Solution:
    def maxArea(self, height):
        height = np.array(height)
        order = height.copy()
        order = list(set(order))
        order.sort(reverse=True)
        maximum = 0
        if sum(height == order[0]) < 2 :
            weight = np.where(height == max(height))[0]
            height[weight] = order[1]

        for index ,i in enumerate(order):
            weight = np.where(height == i)[0]
            water = (max(weight) - min(weight))*i
            if maximum < water:
                maximum = water
            if i != min(order):
                height[weight] = i - 1 
            else :
                return maximum
```

###  參考解答後的答案

```python
class Solution:
    def maxArea(self, height):
        
        left,maximum,right  = 0,0,len(height)-1
        
        for i in range(len(height)-1):
            temp = min(height[right],height[left]) * (len(height)-i-1)
            maximum = max(maximum,temp)
            if height[left] > height[right] :
                right -= 1
            else :
                left += 1
        return maximum
```

###  修改後成績更好的答案

```python
class Solution:
    def maxArea(self, height):
        
        left,maximum,right  = 0,0,len(height)-1
        
        for i in range(len(height)-1 , 0 , -1):
            maximum = max( maximum , min(height[right],height[left])*i )
            if height[left] > height[right] :
                right -= 1
            else :
                left += 1
        return maximum
```
