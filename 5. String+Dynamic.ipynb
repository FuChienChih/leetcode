#### 動態規劃Dynamic Programming
- 動態規劃是分治法的延伸。當分治法分割出來的問題，一而再、再而三出現，就運用記憶法儲存這些問題的答案，避免重複求解，以空間換取時間。

```python
class Solution:
    def longestPalindrome(self, s):
        self.res = ''
        for i in range(len(s)):
            l = n = i
            self.dfs(l,n,s) # 對稱軸是一個字元的時候
            if n+1 <= len(s)-1 and s[l] == s[n+1]:
                self.dfs(l,n+1,s) # 對稱軸是兩個字元的時候
        return self.res
    def dfs(self,l,n,s) :
        # 如果下一組還是對稱，則進入下一個遞迴。否則判斷這層的len是否最大。
        if 0 <= l-1 and n+1 <= len(s)-1 and s[l-1] == s[n+1]:
            self.dfs(l-1,n+1,s)
        else:
            if len(s[l:n+1]) > len(self.res):
                self.res = s[l:n+1]
```

```python
a = Solution()
a.longestPalindrome('aabacabcaa')
```
