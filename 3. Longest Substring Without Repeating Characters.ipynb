```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        temp = 0 
        substr = ''
        for i in s:
            if i not in substr:
                substr += i
                continue
            temp = max(temp,len(substr))
            substr = substr[substr.index(i)+1:]+i
        return max(temp,len(substr))
```
