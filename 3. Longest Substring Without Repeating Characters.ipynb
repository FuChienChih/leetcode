# [解答]

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        temp = 0 
        substr = ''
        for i in s:
            if i not in substr:
                substr += i
                continue
            temp = max(temp,len(substr))
            substr = substr[substr.index(i)+1:]+i
        return max(temp,len(substr))
```

# [解題]

- substr : 儲存連續字母 (預設 = ' ')
- temp : 儲存substr(最大)長度
- 例：'abcacbb'
    - 迴圈判斷字母 i 是否存在substr中
        - 否： substr + i
        - 是： temp = max(temp,len(substr)) 。 將字母i(包含)之前的所有字母從substr中移除，並將 i 從新加入substr。

```python
s = 'abhafnadsajdno'
temp = 0 
substr = ''
for i in s:
    if i not in substr:
        substr += i
        continue
    temp = max(temp,len(substr))
    substr = substr[substr.index(i)+1:]+i
max(temp,len(substr))
```
