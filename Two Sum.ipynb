Given an array of integers nums and an integer target, return indices of the two numbers such 

that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

nums --> list

target --> int 

return --> list[int]

### 最初解法：

```python
class Solution:
    def twoSum(self,nums,target):
        for i,num in enumerate(nums):
            dis = target -num 
            if dis in nums and i != nums.index(dis):
                return[i,nums.index(dis)]
```

### 查詢了題目的意義，發現題目想讓我們思考，如何使用空間儉省時間。(hash table)

```python
class Solution:
    def twoSum(self,nums, target):
        dic = {}
        for i in range(len(nums)) :
            if target-nums[i] in dic:
                return [dic[target-nums[i]],i]
            dic[nums[i]] = i
        return []
```

### 測試

```python
import time

nums_list = [[1,2,3],[3, 2, 4],[2,3,5],[7,1,2,6]]

target_list =  [3,6,7,7]

ans_list = [[0,1],[1,2],[0,2],[1,3]]
```

```python
time_start = time.time()# 計時開始

# 主程式
solution = Solution()
for nums,target,ans in zip(nums_list,target_list,ans_list):
    print(ans == solution.twoSum(nums , target))
    
time_end = time.time()# 計時結束
print('\ntime cost', time_start - time_end, 's')
```
