Given an array of integers nums and an integer target, return indices of the two numbers such 

that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

nums --> list

target --> int 

return --> list[int]

知識點：

    這種兩數和的問題，我們直覺會想到『哪兩個數相加 會等於target』，但對於程式來說，它需要將一個list中的每兩個都加過一遍，才能得到解答。相當於握手問題，需要的運算較多。
    所以我們要想的是，如何減少運算-->我們可以利用target減去第一個數，並觀查這個值是否在list中就好。
    

```python
nums = [3, 2, 4]
target = 6
```

```python
class Solution:
    def twoSum(self,nums, target):
        for num in nums:
            dis = target - num
            if dis in nums and nums.index(num) != nums.index(dis):
                return [nums.index(num) , nums.index(dis)]
```

```python
a = Solution()
a.twoSum(nums , target)
```
