### 圖解，以candidates=[2,3,5] , target = 8 為例

```python
class Solution():
    def combinationSum(self, candidates, target):
        candidates.sort()
        self.candidates = candidates
        self.res = []
        self.dfs([],target,0)
        return self.res
            
    def dfs(self,sublist,target,last):
        if target == 0:
            self.res.append(sublist) 
        if target < self.candidates[0]: 
            return 
        for num in self.candidates:
            if num > target:
                return 
            if num < last:
                continue
            self.dfs(sublist+[num],target-num,num)
```
