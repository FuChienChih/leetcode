```python
class Solution:
    def generateParenthesis(self, n):
        self.res = []
        self.dfs(n,n,'')
        return self.res
    def dfs(self,l,r,temp):
        if l == r == 0 :
            self.res.append(temp)
        if l > 0 :
            self.dfs(l-1,r,temp+'(')
        if r > l :
            self.dfs(l,r-1,temp+')')
```
